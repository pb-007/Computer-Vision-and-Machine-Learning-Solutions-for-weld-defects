```python
zf = zipfile.ZipFile(io.BytesIO(uploaded['defective.zip']), "r")
zf.extractall()
os.listdir('.')
```

```python
images = [os.path.join('defective/images', x) for x in os.listdir('defective/images')]
annotations = [os.path.join('defective/annotations', x) for x in os.listdir('defective/annotations') if x[-3:] == "txt"]

images.sort()
annotations.sort()

# Split the dataset into train-valid-test splits 
train_images, val_images, train_annotations, val_annotations = train_test_split(images, annotations, test_size = 0.1, random_state = 1)
val_images, test_images, val_annotations, test_annotations = train_test_split(val_images, val_annotations, test_size = 0.5, random_state = 1)
```

```python
!mkdir defective/images/train defective/images/val defective/images/test defective/annotations/train defective/annotations/val defective/annotations/test
```

```python
def move_files_to_folder(list_of_files, destination_folder):
    for f in list_of_files:
        try:
            shutil.move(f, destination_folder)
        except:
            print(f)
            assert False

# Move the splits into their folders
move_files_to_folder(train_images, 'defective/images/train')
move_files_to_folder(val_images, 'defective/images/val/')
move_files_to_folder(test_images, 'defective/images/test/')
move_files_to_folder(train_annotations, 'defective/annotations/train/')
move_files_to_folder(val_annotations, 'defective/annotations/val/')
move_files_to_folder(test_annotations, 'defective/annotations/test/')
```
